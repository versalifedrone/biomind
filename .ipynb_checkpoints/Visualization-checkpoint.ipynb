```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
trials = pd.DataFrame({'Disorder': ['Depression', 'Schizophrena', 'Anxiety', 
                                   'Hallucinations', 'Dysthymia', 'Sleep', 
                                   'Appetite', 'Neuro-endocrine', 'Migraine'],
                     'Trials': [8, 1, 2, 0, 1, 0, 0, 4, 4]})
trials =trials.sort_values(by=['Trials'])

consulations = pd.DataFrame({'Year': [2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017],
                       'Medical consulations': [item/38929489*100 for item in [38929489,39469831,40727703,40963440,40792751,41588137,
                                                40776825,40059714,39941172,40139464,40542347,40912393]],
                       'Psychiatric consulations': [item/232865*100 for item in [232865,241242,300921,383438,409269,433301,
                                                    478133,474531,459814,465053,467860,437750]]
                      })
```

```python
drugs = pd.read_csv("depresso.csv")
drugs.loc[:, "Year"] = drugs.loc[:, "Year"].astype(int)
```
